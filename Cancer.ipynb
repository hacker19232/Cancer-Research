```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Cancer_Data.csv")
df.head()
df.columns
```

```python
df
```

```python
#cleaning data
print("Number of NaNs", df.isna().sum().sum())
df.drop(columns=['Unnamed: 32', 'id'], inplace=True)
df.dropna(inplace=True)
print("Number of NaNs", df.isna().sum().sum())
df["diagnosis"] = np.where(df["diagnosis"]=='M', 0,1)
print("Unique Values in Diagnosis Type", df['diagnosis'].unique())
df['diagnosis']
df.drop(index=1, inplace = True)
#test = list(df.loc[0])

```

```python
y = df["diagnosis"]
X = df
X.drop(columns=["diagnosis"], inplace=True)
```

```python
y, X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model = model.fit(X_train, y_train)

```

```python
y_pred = model.predict(X_test)
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(results.shape)
print(results["Predicted"].unique)
```

```python
#accuracy score
accuracy = metrics.accuracy_score(y_test, y_pred)
print('The accuracy is', round(accuracy*100), "%")
```
